```python
!pip install pandas numpy matplotlib seaborn scikit-learn nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, confusion_matrix
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import re
import nltk
```

```python
# Этап 1: Загрузка датасета
from sklearn.datasets import fetch_20newsgroups

categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']
newsgroups = fetch_20newsgroups(subset='all', categories=categories, remove=('headers', 'footers', 'quotes'))

df = pd.DataFrame({'text': newsgroups.data, 'category': newsgroups.target})
df['category_name'] = df['category'].apply(lambda x: newsgroups.target_names[x])

print(f"Всего документов: {len(df)}")
print("\nРаспределение категорий:")
print(df['category_name'].value_counts())
```

```python
# Этап 2: Предобработка текста
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    if not isinstance(text, str):
        return ""

    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    try:
        tokens = word_tokenize(text)
    except LookupError:

        tokens = text.split()

    tokens = [lemmatizer.lemmatize(word) for word in tokens
              if word not in stop_words and len(word) > 2]
    return ' '.join(tokens)

df['processed_text'] = df['text'].apply(preprocess_text)

print("\nПример обработанного текста:")
print(df['processed_text'].iloc[0])
```

```python
# Этап 3: Анализ данных
# Длина текстов до и после обработки
df['original_length'] = df['text'].apply(len)
df['processed_length'] = df['processed_text'].apply(len)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['original_length', 'processed_length']])
plt.title('Длина текстов до и после обработки')
plt.show()

# Распределение по категориям
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='category_name')
plt.title('Распределение документов по категориям')
plt.show()
```

```python
# Этап 4: Векторизация текста
tfidf = TfidfVectorizer(max_features=5000, min_df=5, max_df=0.7)
X = tfidf.fit_transform(df['processed_text'])
y = df['category']

print(f"\nРазмерность матрицы признаков: {X.shape}")
```

```python
# Этап 5: Кластеризация и сравнение с реальными категориями
# Кластеризация K-means
kmeans = KMeans(n_clusters=len(categories), random_state=42)
clusters = kmeans.fit_predict(X)

# Добавляем кластеры в DataFrame
df['cluster'] = clusters

# Оценка качества кластеризации
ari = adjusted_rand_score(y, clusters)
print(f"Adjusted Rand Index: {ari:.3f}")

# Матрица путаницы
conf_matrix = confusion_matrix(y, clusters)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d',
            xticklabels=['Cluster '+str(i) for i in range(len(categories))],
            yticklabels=[newsgroups.target_names[i] for i in range(len(categories))])
plt.title('Матрица путаницы: реальные категории vs кластеры')
plt.xlabel('Кластеры')
plt.ylabel('Реальные категории')
plt.show()
```

```python

# Этап 6: Разделение данных на train, test и validation
# Сначала разделяем на train+val и test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y)

# Затем разделяем train+val на train и val
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.15/0.85, random_state=42, stratify=y_temp)

# Проверяем размеры выборок
print("\nРазмеры выборок:")
print(f"Train: {X_train.shape[0]} samples")
print(f"Validation: {X_val.shape[0]} samples")
print(f"Test: {X_test.shape[0]} samples")

# Проверяем распределение классов в выборках
print("\nРаспределение классов в train:")
print(pd.Series(y_train).value_counts(normalize=True))
print("\nРаспределение классов в validation:")
print(pd.Series(y_val).value_counts(normalize=True))
print("\nРаспределение классов в test:")
print(pd.Series(y_test).value_counts(normalize=True))
```
